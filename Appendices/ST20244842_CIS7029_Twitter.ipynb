```python
#copying data frame to new variable
df = tweets_df2.copy()
df.head()
```

```python
import re
from nltk import PorterStemmer
```

```python
#clean tweets by removing tags, urls etc, and creating another column to place clean tweets
df['Tidy_Tweet'] = df['Tweet'].str.replace(r"([^0-9A-Za-z \t])|(\w+:\/\/\S+)", ' ')
```

```python
#delete all words with length less than 3 letters
df['Tidy_Tweet'] = df['Tidy_Tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))

#tokenize tweet to separate all words
tokenized_tweet = df['Tidy_Tweet'].apply(lambda x: x.split())
```

```python
#stem and apply lower case to the tweets
ps = PorterStemmer()
tokenized_tweet = tokenized_tweet.apply(lambda x: [ps.stem(i) for i in x])
```

```python
#remove tokenization
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = ' '.join(tokenized_tweet[i])

df['Tidy_Tweet'] = tokenized_tweet
```

```python
from textblob import TextBlob
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import numpy as np
import urllib
import requests
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#compute sentiments of each tweet
sentiment_objects = [TextBlob(tweet) for tweet in df['Tidy_Tweet']]
sentiment_values = [[tweet.sentiment.polarity, (tweet)] for tweet in sentiment_objects]

#assign sentiments to dataframe
sentiment_df = pd.DataFrame(sentiment_values, columns=["polarity", "tweet"])
df['Polarity'] = sentiment_df['polarity']
df.head()
```

```python
#describe sentiments in words and put in a new column
sentiment_sign = sentiment_df["polarity"].apply(np.sign).replace({-1: 'Negative', 1: 'Positive', 0: 'zero'})
df["sentiment_sign"]=sentiment_sign
df.head()
```

```python
#plot a word cloud of positive sentiments
all_words_positive = ' '.join(text for text in df['Tidy_Tweet'][df['sentiment_sign']=="Positive"])

Mask = np.array(Image.open(requests.get('http://clipart-library.com/image_gallery2/Twitter-PNG-Image.png', stream=True).raw))
image_colors = ImageColorGenerator(Mask)

wc = WordCloud(background_color='black', height=1500, width=4000,mask=Mask).generate(all_words_positive)

plt.figure(figsize=(10,20))
plt.imshow(wc.recolor(color_func=image_colors),interpolation="hamming")
plt.axis('off')
plt.show()
```

```python
#plot a word cloud of negative sentiments
all_words_negative = ' '.join(text for text in df['Tidy_Tweet'][df['sentiment_sign']=="Negative"])

Mask = np.array(Image.open(requests.get('http://clipart-library.com/image_gallery2/Twitter-PNG-Image.png', stream=True).raw))
image_colors = ImageColorGenerator(Mask)

wc = WordCloud(background_color='brown', height=1500, width=4000,mask=Mask).generate(all_words_negative)

plt.figure(figsize=(10,20))
plt.imshow(wc.recolor(color_func=image_colors),interpolation="gaussian")
plt.axis('off')
plt.show()
```

```python
#plot a word cloud of neutral sentiments
all_words_neutral = ' '.join(text for text in df['Tidy_Tweet'][df['sentiment_sign']=="zero"])

Mask = np.array(Image.open(requests.get('http://clipart-library.com/image_gallery2/Twitter-PNG-Image.png', stream=True).raw))
image_colors = ImageColorGenerator(Mask)

wc = WordCloud(background_color='white', height=1500, width=4000,mask=Mask).generate(all_words_neutral)

plt.figure(figsize=(10,20))
plt.imshow(wc.recolor(color_func=image_colors),interpolation="hamming")
plt.axis('off')
plt.show()
```
